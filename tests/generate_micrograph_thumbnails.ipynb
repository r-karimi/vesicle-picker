```python
# Brief notebook to experiment with different filters to generate a nice figure for CEC's manuscript
from cv2 import GaussianBlur, medianBlur, bilateralFilter
from vesicle_picker import helpers
import matplotlib.pyplot as plt
```

```python
# Import the micrograph
image, header = helpers.import_mrc("FoilHole_11264182_Data_10178671_10178673_20231011_220254_EER_patch_aligned_doseweighted.mrc")
```

```python
# Show the raw image
plt.imshow(image, cmap="Greys_r")
```

```python
# Do a Gaussian blur
gaussian = GaussianBlur(image, (21, 21), 3, 3)
plt.imshow(gaussian, cmap="Greys_r")
plt.axis('off')
plt.tight_layout()
plt.savefig('example_micrograph_gaussian.png', dpi=600, bbox_inches="tight", pad_inches=0)
plt.show()
```

```python
# Do a bilateral filter
bilateral = bilateralFilter(image, 21, 101, 101)
plt.imshow(bilateral, cmap="Greys_r")
plt.axis('off')
plt.tight_layout()
plt.savefig('example_micrograph_bilateral.png', dpi=600, bbox_inches="tight", pad_inches=0)
plt.show()
```

```python
# Show the micrographs side by side
fig, axes = plt.subplots(1, 3, figsize=(16, 8))

axes[0].imshow(image, cmap="Greys_r")
axes[1].imshow(gaussian, cmap="Greys_r")
axes[2].imshow(bilateral, cmap="Greys_r")

plt.show()
```
